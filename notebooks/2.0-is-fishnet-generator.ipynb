# Fishnet Generator
Let's define a function that creates a fishnet grid for an input GeoDataFrame. The source code from `build_fishnet.py` was based off of https://spatial-dev.guru/2022/05/22/create-fishnet-grid-using-geopandas-and-shapely/

```python
# geopandas imports
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely import geometry

# directory
import os
current_dir = os.path.dirname("")
interim_shp_path = os.path.join(current_dir, "../data/interim/shp/")
spatial_path = data_path = os.path.join(current_dir, "../spatial/build_fishnet.py")

#sys
import sys
sys.path.append("../spatial/")

import build_fishnet

chicago_gdf = gpd.read_file(
    os.path.join(
        interim_shp_path, "chicago/chicago_epsg3435.shp"
    )
)

example_fishnet = build_fishnet.build_fishnet(chicago_gdf, 500)

# plot results
fig, ax = plt.subplots(figsize = (10,10))
example_fishnet.boundary.plot(linewidth = 0.5, ax=ax)
chicago_gdf.boundary.plot(ax = ax)
plt.show()
```

As you can see the fishnet takes a conservative approach to generation. The `build_fishnet()` function creates a naive boundary along the extent of the input shapefile and filters polygons in the fishnet that intersects with the input shapefile.
